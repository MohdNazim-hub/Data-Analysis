### Weighted Hybrid Technique

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
credits=pd.read_csv("E:/csv/tmdb_5000_credits.csv")
movies=pd.read_csv("E:/csv/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head(2)
```

```python
#need to merge both the dataframes
# can be done by using movie_id
```

```python
movies=movies.rename(columns={'id':'movie_id'})
movies.head(2)
```

```python
#merging through movie_id
new=pd.merge(movies,credits,on='movie_id')
new.head(2)
```

```python
#dropping unnnecessary columns
new1=new.drop(['homepage','keywords','original_language','overview',
               'production_companies','spoken_languages','status','tagline',
               'title_x','title_y','crew'],axis=1)
new1.head(2)
```

```python
new1.describe()
```

```python
new1.info()#no null values
```

Weighted Average Of each movie rating: 
W=(R*v+C*m)/(v+m)

R=average movie rating

v=no of votes for movie

m=minimum rating to be listed in chart

C=mean votes in whole report

```python
R=new1['vote_average']
v=new1['vote_count']
C=new1['vote_average'].mean()
m=new1['vote_count'].quantile(.90)#to be in chart it has to be votes more than 90%
```

```python
new1['weighted_average']=(R*v+C*m)/(v+m)
```

```python
new1.head(2)
```

```python
new2=new1[['movie_id','original_title','popularity','vote_average','vote_count','weighted_average']]
final=new2.sort_values(by=['weighted_average'],ascending=False)
final.head()
```

```python
final1=final.iloc[0:10,:]
final1
```

```python
plt.pie(final1['weighted_average'],labels=final1.original_title,autopct='%0.1f%%',explode=(0.3,0,0,0,0,0,0,0,0,0),radius=2)
plt.show()
```

```python
plt.barh(final1.original_title,final1.popularity)
plt.show()
```
