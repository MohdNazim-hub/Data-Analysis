```python
#make a database named sqlproject
#make a table accounts(username varchar(20),age int,mobile_no bigint,address varchar(20),password varchar(20))
from tkinter import *
import mysql.connector as con
from tkinter import messagebox

root=Tk()
root.geometry('400x400+200+150')
root.title('LoginPage')

uservar=StringVar()
agevar=StringVar()
mobvar=StringVar()
addvar=StringVar()
passvar=StringVar()

l0=Label(root,font=('ArialBold',14),text='LOGIN/SIGNUP')
l0.place(x=8,y=20)

l1=Label(root,font=('ArialBold',14),text='Username :')
l1.place(x=30,y=100)
e1=Entry(root,bd=3,font=("ArialBold",13),textvariable=uservar)
e1.place(x=150,y=100,height=27,width=180)

l2=Label(root,font=('ArialBold',14),text='Password :')
l2.place(x=30,y=140)
e2=Entry(root,bd=3,font=("ArialBold",13),textvariable=passvar)
e2.place(x=150,y=140,height=27,width=180)



def signup():
    signup=Toplevel(root)
    signup.title("SignupPage")
    signup.geometry("400x400+200+150")

    l0=Label(signup,font=('ArialBold',14),text='SIGNUP')
    l0.place(x=8,y=30)

    l1=Label(signup,font=('ArialBold',14),text='Username :')
    l1.place(x=30,y=100)
    e1=Entry(signup,bd=3,font=("ArialBold",13),textvariable=uservar)
    e1.place(x=180,y=100,height=27,width=180)

    l2=Label(signup,font=('ArialBold',14),text='Age :')
    l2.place(x=30,y=140)
    e2=Entry(signup,bd=3,font=("ArialBold",13),textvariable=agevar)
    e2.place(x=180,y=140,height=27,width=180)

    l3=Label(signup,font=('ArialBold',14),text='Mobile Number :')
    l3.place(x=30,y=180)
    e3=Entry(signup,bd=3,font=("ArialBold",13),textvariable=mobvar)
    e3.place(x=180,y=180,height=27,width=180)

    l4=Label(signup,font=('ArialBold',14),text='Address :')
    l4.place(x=30,y=220)
    e4=Entry(signup,bd=3,font=("ArialBold",13),textvariable=addvar)
    e4.place(x=180,y=220,height=27,width=180)

    l5=Label(signup,font=('ArialBold',14),text='Password :')
    l5.place(x=30,y=260)
    e5=Entry(signup,bd=3,font=("ArialBold",13),textvariable=passvar)
    e5.place(x=180,y=260,height=27,width=180)
    def submit():
        user=uservar.get()
        age=agevar.get()
        mob=mobvar.get()
        add=addvar.get()
        pass_=passvar.get()
        x=list(pass_)
        num=0
        alpha=0
        special=0
        for i in x:
            if i in ('1','2','3','4','5','6','7','8','9','0'):
                num+=1 
            if i in ('!','@','$','#','^','&','*','(',')','?','+'):
                special+=1
            if i in list('qwertyuiopasdfghjklzxcvbnm'):
                alpha+=1
        if user!='' and pass_!='':
            if num>0 and alpha>1 and special>0:
                connect=con.connect(host='localhost',
                                user='root',
                                password='',
                                database='sqlproject')
                connect
                cursor=connect.cursor()
                cursor.execute("insert into accounts values('"+user+"',"+age+","+mob+",'"+add+"','"+pass_+"')")
                connect.commit()
                connect.close()
                uservar.set("")
                agevar.set("")
                mobvar.set("")
                addvar.set("")
                passvar.set("")
                messagebox.showinfo('Information','Data is Saved')
            else:
                messagebox.showinfo('Information','password should have \n 1 numeric value\n 2 character value\n and special symbol')
        else:
            messagebox.showinfo('Information','Enter details')

    b1=Button(signup,fg='white',command=submit,bg='grey',font=("ArialBold",13),text="Submit",bd=3)
    b1.place(x=250,y=310,width=90)
def login():
    user=uservar.get()
    pass_=passvar.get()
    
    x=list(pass_)
    num=0
    alpha=0
    special=0
    for i in x:
        if i in ('1','2','3','4','5','6','7','8','9','0'):
            num+=1 
        if i in ('!','@','$','#','^','&','*','(',')','?','+'):
            special+=1
        if i in list('qwertyuiopasdfghjklzxcvbnm'):
            alpha+=1
       
    
    connect=con.connect(host='localhost',
                    user='root',
                    password='',
                    database='sqlproject')
    connect
    cursor=connect.cursor()
    cursor.execute("select * from accounts where username='"+user+"' and password='"+pass_+"'")
    myresult=cursor.fetchone()
    if myresult!=None and num>0 and alpha>1 and special>0:
        login=Toplevel(root)
        login.title("YourAccount")
        login.geometry("400x500+200+50")


        l1=Label(login,font=('ArialBold',14),text='Username :')
        l1.place(x=30,y=240)
        l1_=Label(login,font=('ArialBold',14),text=myresult[0])
        l1_.place(x=140,y=240)
        l2=Label(login,font=('ArialBold',14),text='Age :')
        l2.place(x=30,y=280)
        l2_=Label(login,font=('ArialBold',14),text=myresult[1])
        l2_.place(x=140,y=280)
        l3=Label(login,font=('ArialBold',14),text='Mobile Number :')
        l3.place(x=30,y=320)
        l3_=Label(login,font=('ArialBold',14),text=myresult[2])
        l3_.place(x=190,y=320)        
        l4=Label(login,font=('ArialBold',14),text='Address :')
        l4.place(x=30,y=360)
        l4_=Label(login,font=('ArialBold',14),text=myresult[3])
        l4_.place(x=140,y=360)
    
        
        
        def edit():
            signup=Toplevel(root)
            signup.title("SignupPage")
            signup.geometry("400x400+200+150")

            l0=Label(signup,font=('ArialBold',14),text='Update')
            l0.place(x=8,y=30)

            l1=Label(signup,font=('ArialBold',14),text='Username :')
            l1.place(x=30,y=100)
            e1=Entry(signup,bd=3,font=("ArialBold",13),textvariable=uservar)
            e1.place(x=180,y=100,height=27,width=180)

            l2=Label(signup,font=('ArialBold',14),text='Age :')
            l2.place(x=30,y=140)
            e2=Entry(signup,bd=3,font=("ArialBold",13),textvariable=agevar)
            e2.place(x=180,y=140,height=27,width=180)

            l3=Label(signup,font=('ArialBold',14),text='Mobile Number :')
            l3.place(x=30,y=180)
            e3=Entry(signup,bd=3,font=("ArialBold",13),textvariable=mobvar)
            e3.place(x=180,y=180,height=27,width=180)

            l4=Label(signup,font=('ArialBold',14),text='Address :')
            l4.place(x=30,y=220)
            e4=Entry(signup,bd=3,font=("ArialBold",13),textvariable=addvar)
            e4.place(x=180,y=220,height=27,width=180)

            l5=Label(signup,font=('ArialBold',14),text='Password :')
            l5.place(x=30,y=260)
            e5=Entry(signup,bd=3,font=("ArialBold",13),textvariable=passvar)
            e5.place(x=180,y=260,height=27,width=180)
            def Change():
                age=agevar.get()
                mob=mobvar.get()
                add=addvar.get()
                
                if age!='' and mob!='' and add!='':
                    connect=con.connect(host='localhost',
                                    user='root',
                                    password='',
                                    database='sqlproject')
                    connect
                    cursor=connect.cursor()
                    query1="update accounts set username='"+user+"',age="+age+",mobile_no="+mob+",address='"+add+"', password='"+pass_+"' where username='"+user+"' and password='"+pass_+"'"
                    cursor.execute(query1)
                    connect.commit()
                    signup.destroy()
                    login.destroy()
                    messagebox.showinfo('Information','Data is Updated')
                else:
                    messagebox.showinfo('Information','Enter correct details')
                    
                
            b1=Button(signup,fg='white',command=Change,bg='grey',font=("ArialBold",13),text="Change",bd=3)
            b1.place(x=250,y=310,width=90)

        b1=Button(login,fg='white',bg='grey',command=edit,font=("ArialBold",13),text="Edit Profile",bd=3)
        b1.place(x=47,y=420,width=90)
        
        def delete():
            connect=con.connect(host='localhost',
                    user='root',
                    password='',
                    database='sqlproject')
            connect
            cursor=connect.cursor()
            query="delete from accounts where username='"+user+"'and password='"+pass_+"'"
            cursor=connect.cursor()
            cursor.execute(query)
            connect.commit()
            login.destroy()
            messagebox.showinfo('Information','Data is Deleted')
            
        b1=Button(login,fg='white',bg='grey',command=delete,font=("ArialBold",13),text="Delete Acoount",bd=3)
        b1.place(x=140,y=420)
        
        def close():
            root.destroy()
        b1=Button(login,fg='white',bg='grey',command=close,font=("ArialBold",13),text="Close",bd=3)
        b1.place(x=270,y=420,width=90)

        img =PhotoImage(file="D:/images.png")
        label = Label(login, image=img)
        label.place(x=3,y=3,width=400,height=200)

        login.mainloop()
    else:
        messagebox.showinfo('Information','Not found')
   
b1=Button(root,fg='white',bg='grey',command=login,font=("ArialBold",13),text="Log in",bd=3)
b1.place(x=240,y=203,width=90)


b1=Button(root,fg='white',bg='grey',command=signup,font=("ArialBold",13),text="Sign up",bd=3)
b1.place(x=135,y=203,width=90)

root.mainloop()
```
